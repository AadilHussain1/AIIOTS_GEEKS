```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

```

## Load the dataset

```python
df = pd.read_csv('House Price Prediction Dataset.csv')
```

```python
df.head().style.background_gradient(cmap="PuBu")
```

## Understand the Dataset

```python
from colorama import Fore, Style

# Print number of rows and cols
print(Fore.CYAN + "df shape: " + Style.RESET_ALL)
print(f"{df.shape}\n")
```

```python
# print info
print(Fore.GREEN + "df info: " + Style.RESET_ALL)
print(f"{df.info()}\n")
```

```python
print(Fore.RED + "df Columns: " + Style.RESET_ALL)
print(f"{df.columns.tolist()}\n")
```

```python
# Find the missing (NaN) values
print(Fore.YELLOW + "df Missing Values: " + Style.RESET_ALL)
print(f"{df.isnull().sum()}\n")
```

```python
# Print statitical summary of dataset
print(Fore.MAGENTA + "Statistical Summary: " + Style.RESET_ALL)
print(f"{df.describe()}")
```

```python
# Numerical and Categorical Columns
numerical = df.select_dtypes(include=["int64"]).columns
categorical = df.select_dtypes(include=["object"]).columns

print(f"We are having: {len(numerical)} numerical features {numerical}")

print(80 * "-")

print(f"The categorical features are: {len(categorical)} categorical features {categorical}")
```

# EDA

```python
numerical = ['Area', 'Bedrooms', 'Bathrooms', 'Floors', 'YearBuilt', 'Price']

for col in numerical:
    plt.figure(figsize=(15, 5))
    plt.subplot(1,2,1)
    sb.histplot(data = df, x = col, kde=True, bins = 30, color="cyan")
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

    plt.subplot(1,2,2)
    sb.boxenplot(data = df, x = col, color = "green")
    plt.title(f"Boxplot of {col}")
    plt.tight_layout()
    plt.show()

    print(f"Number of Unique {col}: {df[col].nunique()}")
    print(f"Missing Values in {col}: {df[col].isnull().sum()}")
    
```

# Categorical Features

```python
categorcal = ['Location', 'Condition', 'Garage']

for col in categorical:
    counts = df[col].value_counts()
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sb.countplot(data = df, x = col, palette = "Set2")
    plt.title(f"Count of {col} values")
    plt.xticks(rotation=90)
    plt.ylabel("Count")
    # plt.show()

    plt.subplot(1, 2, 2)
    plt.pie(counts, labels = counts.index, autopct = "%1.1f%%", startangle=90)
    plt.title(f"Percentage of {col}")
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

# Categorical Features vs Label

```python
# Figure with 2 subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 6))

# subplot 1: 
sb.countplot(data = df, y = "Location", order = df["Location"].value_counts().index, hue="Condition",palette="inferno", ax=axs[0])
axs[0].set_title("Count of Location by Condition")
axs[0].set_xlabel("Counts")
axs[0].set_ylabel("Location")
axs[0].tick_params(axis='x', rotation=90)

# subplot 2:
sb.countplot(data = df, y = "Location", order = df["Location"].value_counts().index, hue = "Garage", palette = "inferno", ax=axs[1])
axs[1].set_title("Count of Location by Garage")
axs[1].set_xlabel("Counts")
axs[1].set_ylabel("Location")
axs[1].tick_params(axis='x', rotation=90)

fig.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1,3,1)
sb.barplot(data=df, x="Location", y="Price", hue="Location", palette="Set2")
plt.title("Price by Location")
plt.xlabel("Locations")
plt.ylabel("Price")
plt.grid()

plt.subplot(1,3,2)
sb.barplot(data=df, x="Location", y = "Price", hue="Condition", palette="Set2")
plt.title("Price by Condition")
plt.xlabel("Condition")
plt.ylabel("Price")
plt.grid()

plt.subplot(1, 3, 3)
sb.barplot(data = df, x = "Location", y = "Price", hue = "Garage", palette="Set2")
plt.title("Price by Garage")
plt.xlabel("Garage")
plt.ylabel("Price")
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: Price by Area, colored by Garage
sb.scatterplot(data = df, x = "Area", y = "Price", color = "violet", hue = "Garage")
plt.title("Price by Area (Garage)")
plt.xlabel("Area")
plt.ylabel("Price")
plt.grid()
plt.show()
```

```python
# Scatter plot: Price by Area, colored by Condition
sb.scatterplot(data = df, x = "Area", y = "Price", color = "red", hue = "Condition")
plt.title("Price by Area (Condition) ")
plt.xlabel("Area")
plt.ylabel("Price")
plt.grid()
plt.show()
```

```python
# Scatter plot: Price by Area, colored by Location
sb.scatterplot(data = df, x = "Area", y = "Price", color = "pink", hue = "Location")
plt.title("Price by Area (Location)")
plt.xlabel("Area")
plt.ylabel("Price")
plt.grid()
plt.show()
```

```python
correlation = df[numerical].corr()
plt.figure(figsize=(10, 8))
sb.heatmap(correlation, annot = True, fmt = ".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
