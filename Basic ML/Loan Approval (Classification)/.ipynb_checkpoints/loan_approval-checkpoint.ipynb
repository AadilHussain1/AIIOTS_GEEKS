# EDA

```python
numerical = ['no_of_dependents', 'income_annum', 'loan_amount', 'loan_term', 'cibil_score', 'residential_assets_value','commercial_assets_value', 'luxury_assets_value', 'bank_asset_value']
for col in numerical:
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    sb.histplot(data = df, x = col, kde = True, bins = 30, color = "magenta")
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

    plt.subplot(1,2,2)
    sb.boxenplot(data = df, x = col, color = "blue")
    plt.title(f"Boxplot of {col}")
    plt.tight_layout()
    plt.show()

    # print(f"Numbre of Unique {col}: {df[col].nunique()}")
    # print(f"Missing Values in {col}: {df[col].isnull()}")
```

```python
categorical = ['education', 'self_employed', 'loan_status']

for col in categorical:
    counts = df[col].value_counts()
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sb.countplot(data = df, x = col, hue = col,  palette = "Set2")
    plt.title(f"Count of {col} values")
    plt.xticks(rotation = 90)
    plt.ylabel("Count")

    plt.subplot(1, 2, 2)
    plt.pie(counts, labels = counts.index, autopct = "1%.1f%%", startangle = 90)
    plt.title(f"Percemtage of {col}")
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```
